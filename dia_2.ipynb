```python
# Challenge 4 (manejo de lista)
def agregar_elemento(lista_nombres:list):
    print (lista_nombres)
    nuevo_nombre= input ("Ingrese el nombre nuevo nombre: ")
    lista_nombres.append (nuevo_nombre)
    print (f"Se agrego el nombre: {nuevo_nombre}")
    print (f"La nueva lista es: \n{lista_nombres}")

def modificar_elemento(lista_nombres:list):
    indice_modificar = int (input("Ingrese el indice a modificar: "))
    lista_nombres[indice_modificar-1]= input("Ingrese el nuevo nombre: ")
    print (f"La nueva lista es: \n{lista_nombres}")

def eliminar_porNombre (lista_nombres:list):
    eliminar_nombre = input("Ingrese el nombre a eliminar: ").strip()
    print (f"El nombre eliminado es: {eliminar_nombre}")
    lista_nombres.remove(eliminar_nombre)
    print (lista_nombres)

def eliminar_porIndice(lista_nombres:list):
    print (f"Cual es el nombre que quieres eliminar: \n {lista_nombres}")
    indice_eliminado= int(input("Ingrese el indice: "))
    # if indice_eliminado> (len(lista_nombres)+1):
    #     print ("Error ")
    nombre_eliminado=lista_nombres[indice_eliminado-1]
    print (f"El nombre que se va a eliminar es: {nombre_eliminado}")
    lista_nombres.pop (indice_eliminado-1)
    print (f"La nueva lista es: \n{lista_nombres}")


nombres = ["Lucas", "Andrea", "Benjamin", "Fatima"]

agregar_elemento(nombres)
modificar_elemento (nombres)
eliminar_porNombre (nombres)
eliminar_porIndice (nombres)
```

```python
# Challenge 2 (utilizamos and u or para realizar el ejercicios)

def verificar_temperatura (temperatura):
    print (f"La temperatura actual es {temperatura}°C")
    if -10<=temperatura<=13 : # temperatura>=-10 and temperatura<=13
        print ("Ro'y koko\n")

    elif 14<=temperatura<=23: # temperatura>=14 and temperatura <=23
        print ("Frio del mono\n")

    elif 24<=temperatura<=28: # temperatura>=24 and temperatura <=28
        print ("Oikoite\n")

    elif 29<=temperatura<=45: # temperatura>=29 and temperatura <=45
        print ("Hakurooo\n")

    else:
        print ("Ñamano Mba'e\n")   


temp_actual= int (input("Ingresar la temperatura actual en °C: "))


verificar_temperatura (temp_actual)
verificar_temperatura (-11)
verificar_temperatura (12)
verificar_temperatura (14)
verificar_temperatura (23)
verificar_temperatura (27)
verificar_temperatura (45)
verificar_temperatura (50)

```
