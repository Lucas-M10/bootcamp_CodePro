```python
#Challenge 4 
cant_caramelos = int (input("Ingrese el total de caramelos: "))
cant_ninhos = int (input("Ingrese la cantidad de niños: "))
# Cuanto recibe cada ninho
caramelos_recib = cant_caramelos//cant_ninhos
# Cuantos caramelos sobra 
caramelos_restantes = cant_caramelos%cant_ninhos
print (f"""Son {cant_ninhos} niños y hay {cant_caramelos}
A cada niño le toca {caramelos_recib}
Sobraron {caramelos_restantes}""")
```
